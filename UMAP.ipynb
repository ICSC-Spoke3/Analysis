```python
%time
trans = umap.UMAP(n_neighbors=10, random_state=42).fit(X_train_scaled)
embedding = trans.transform(X_train_scaled)
```

```python
param_grid_xgb = {
    'max_depth': [3],#, 5, 6],6 e' il max
    'learning_rate': [0.1],# 0.01, 0.001],
    'subsample': [0.5],# 0.7, 1],
   # 'reg_alpha':[0],# 0.5, 1, 5],
    #'reg_lambda':[0],# 0.5, 1, 5]   
}

xgb_model = xgb.XGBClassifier(tree_method="hist",seed=0)
grid_search = GridSearchCV(xgb_model, param_grid_xgb, cv=5, scoring='accuracy',refit=True)
# Fit the GridSearchCV object to the training data
X_train, X_test, y_train, y_test = train_test_split(shuffled_df.iloc[:, 1:-1], shuffled_df.iloc[:, -1], test_size=0.3, random_state=0)
X_val, X_test1, Y_val, Y_test1 = train_test_split(X_test, y_test, test_size=0.1, random_state=0)
grid_search.fit(embedding, y_train)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test1_scaled = scaler.transform(X_test1)

# Riduzione dimensionale con UMAP sul set di addestramento
reducer = umap.UMAP(n_neighbors=10, random_state=42)
embedding_train = reducer.fit_transform(X_train_scaled)
embedding_val = reducer.transform(X_val_scaled)
embedding_test1 = reducer.transform(X_test1_scaled)


# Eseguire la ricerca a griglia sul set di addestramento ridotto
grid_search.fit(embedding_train, y_train)

# Ottenere il modello con i migliori parametri
best_model = grid_search.best_estimator_

# Predizione e valutazione sul set di validazione
val_predictions = best_model.predict(embedding_val)
val_accuracy = accuracy_score(Y_val, val_predictions)
print(f"Validation Accuracy: {val_accuracy}")

# Predizione e valutazione sul set di test finale
test_predictions = best_model.predict(embedding_test1)
test_accuracy = accuracy_score(Y_test1, test_predictions)
print(f"Test Accuracy: {test_accuracy}")
```
